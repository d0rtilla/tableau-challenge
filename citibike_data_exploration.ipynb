```python
import pandas as pd
from datetime import datetime
import numpy as np 
```

```python
# Load csv files

feb2021_url = 'JC-202102-citibike-tripdata.csv'
feb2021_df = pd.read_csv(feb2021_url)


mar2021_url = 'JC-202103-citibike-tripdata.csv'
mar2021_df = pd.read_csv(mar2021_url)


apr2021_url = 'JC-202104-citibike-tripdata.csv'
apr2021_df = pd.read_csv(apr2021_url)


may2021_url = 'JC-202105-citibike-tripdata.csv'
may2021_df = pd.read_csv(may2021_url)


jun2021_url = 'JC-202106-citibike-tripdata.csv'
jun2021_df = pd.read_csv(jun2021_url)


jul2021_url = 'JC-202107-citibike-tripdata.csv'
jul2021_df = pd.read_csv(jul2021_url)


aug2021_url = 'JC-202108-citibike-tripdata.csv'
aug2021_df = pd.read_csv(aug2021_url)


sep2021_url = 'JC-202109-citibike-tripdata.csv'
sep2021_df = pd.read_csv(sep2021_url)


oct2021_url = 'JC-202110-citibike-tripdata.csv'
oct2021_df = pd.read_csv(oct2021_url)


nov2021_url = 'JC-202111-citibike-tripdata.csv'
nov2021_df = pd.read_csv(nov2021_url)


dec2021_url = 'JC-202112-citibike-tripdata.csv'
dec2021_df = pd.read_csv(dec2021_url)

jan2022_url = 'JC-202201-citibike-tripdata.csv'
jan2022_df = pd.read_csv(jan2022_url)

feb2022_url = 'JC-202202-citibike-tripdata.csv'
feb2022_df = pd.read_csv(feb2022_url)


mar2022_url = 'JC-202203-citibike-tripdata.csv'
mar2022_df = pd.read_csv(mar2022_url)


apr2022_url = 'JC-202204-citibike-tripdata.csv'
apr2022_df = pd.read_csv(apr2022_url)


may2022_url = 'JC-202205-citibike-tripdata.csv'
may2022_df = pd.read_csv(may2022_url)


jun2022_url = 'JC-202206-citibike-tripdata.csv'
jun2022_df = pd.read_csv(jun2022_url)


jul2022_url = 'JC-202207-citbike-tripdata.csv'
jul2022_df = pd.read_csv(jul2022_url)


aug2022_url = 'JC-202208-citibike-tripdata.csv'
aug2022_df = pd.read_csv(aug2022_url)


sep2022_url = 'JC-202209-citibike-tripdata.csv'
sep2022_df = pd.read_csv(sep2022_url)
```

```python
# Concatenate dataframes into one
data_df=pd.concat([feb2021_df, mar2021_df, apr2021_df, may2021_df, jun2021_df, jul2021_df, aug2021_df, sep2021_df, oct2021_df, nov2021_df, dec2021_df, jan2022_df, feb2022_df, mar2022_df, apr2022_df, may2022_df, jun2022_df, jul2022_df, aug2022_df, sep2022_df])
```

```python
# Drop null values
data_df.dropna()
```

```python
f = lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
data_df.started_at = data_df.started_at.apply(f)
data_df.ended_at = data_df.ended_at.apply(f)
```

```python
# Create column for ride duration
data_df['ride_duration'] = data_df.ended_at - data_df.started_at
data_df
```
